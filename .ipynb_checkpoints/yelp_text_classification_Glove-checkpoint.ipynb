```python
import tensorflow as tf
from keras.backend.tensorflow_backend import set_session
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
config.log_device_placement = True
sess = tf.Session(config=config)
set_session(sess)
```

```python
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense, Dropout
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.callbacks import EarlyStopping, ModelCheckpoint
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import re
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from sklearn.manifold import TSNE
%matplotlib inline
```

```python
df = pd.read_csv('yelp.csv')
df.head()
# nltk.download('stopwords')
```

```python
# drop all rows that have missing values
df.dropna() 
df = df[['text', 'stars']]
df.head()
```

```python
labels = df['stars'].map(lambda x : 2 if int(x) > 3 else 1 if int(x)==3 else 0)
# labelSS = df['stars'].map(lambda x : 2 if int(x) > 3 else 1 if int(x)==3 else 0)
labels[0]
```

```python
def clean_text(text):
#     remove punctuation
    text = text.translate(string.punctuation)
#     convert to lowercase
    text = text.lower().split()
#     remove stop words
    stops = set(stopwords.words('english'))
    text = [w for w in text if not w in stops and len(w) >= 3]
    text= " ".join(text)
    
#     clean the review text
    text = re.sub(r"[^A-Za-z0-9^,!.\/'+-=]", " ", text)
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r",", " ", text)
    text = re.sub(r"\.", " ", text)
    text = re.sub(r"!", " ! ", text)
    text = re.sub(r"\/", " ", text)
    text = re.sub(r"\^", " ^ ", text)
    text = re.sub(r"\+", " + ", text)
    text = re.sub(r"\-", " - ", text)
    text = re.sub(r"\=", " = ", text)
    text = re.sub(r"'", " ", text)
    text = re.sub(r"(\d+)(k)", r"\g<1>000", text)
    text = re.sub(r":", " : ", text)
    text = re.sub(r" e g ", " eg ", text)
    text = re.sub(r" b g ", " bg ", text)
    text = re.sub(r" u s ", " american ", text)
    text = re.sub(r"\0s", "0", text)
    text = re.sub(r" 9 11 ", "911", text)
    text = re.sub(r"e - mail", "email", text)
    text = re.sub(r"j k", "jk", text)
    text = re.sub(r"\s{2,}", " ", text)
#     find the stem of each word
    text = text.split()
    stemmer = SnowballStemmer('english')
    stemmed_words = [stemmer.stem(word) for word in text]
    text = " ".join(stemmed_words)
    
    return text    
```

```python
df['text'] = df['text'].map(lambda x: clean_text(x))
```

```python
df.head(10)
```

```python
maxlen = 50
embed_dim = 100
max_words = 20000
```

```python
tokenzier = Tokenizer(num_words=max_words)
tokenzier.fit_on_texts(df['text'])
```

```python
sequences = tokenzier.texts_to_sequences(df['text'])
data = pad_sequences(sequences, maxlen=50, padding='post')
data[10]
```

```python
vocab_size = len(tokenzier.word_index)+1
vocab_size
# tokenzier.word_index['good']
```

```python
# labels = np.asarray(labels)
print(labels.shape)
print(labels[0])
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
# encode class values as integers
encoder = LabelEncoder()
encoder.fit(labels)
labels = encoder.transform(labels)
# convert integers to dummy variables (i.e. one hot encoded)
labels = np_utils.to_categorical(labels)
print(labels[10:20])
```

```python
from sklearn.model_selection import train_test_split

print(data.shape, labels.shape)
# x_train, x_test, y_train, y_test = train_test_split(data, labelSS, test_size = 0.1, random_state= 0)
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size = 0.1, random_state= 0)
```

```python
# first open the glove text file
embed_index = dict()
f = open('glove.6B/glove.6B.100d.txt', encoding="utf-8")
#  each line in this file belongs to one word
# first element is the word itself and the rest is the vector for that word
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embed_index[word] = coefs
f.close()
```

```python
print(embed_index['apple'])
```

```python
print(len(embed_index)) 
# four hundred thousand words
```

```python
# creating the embedding matrix then adding weights form the Glove to this vector
# but only for words in our current dictionary
embed_matrix = np.zeros((max_words, embed_dim))

for word, i in tokenzier.word_index.items():
    if i < max_words:
        embed_vector = embed_index.get(word)
        if embed_vector is not None:
            embed_matrix[i] = embed_vector
```

```python
from keras.constraints import maxnorm

model = Sequential()
model.add(Embedding(max_words, # 20000
                    embed_dim, #  100
                    weights=[embed_matrix], # get the weights from glove
                    input_length=maxlen))
model.add(Flatten())
model.add(Dropout(0.6))
model.add(Dense(20, activation='relu', kernel_initializer='normal' ,kernel_constraint=maxnorm(3)))
model.add(Dropout(0.6))
model.add(Dense(10, activation='relu', kernel_initializer='normal' ,kernel_constraint=maxnorm(3) ))
model.add(Dropout(0.4))
model.add(Dense(3, activation='softmax'))
# model.summary()
#  changing from sparse categorical cross entropy to cross_entropy didn't show any improvements
# regarding the test accuracy or loss
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
# early_stopping = EarlyStopping(monitor='val_loss', patience=50, mode='min')
save_best = ModelCheckpoint('yelp_comment.hdf', save_best_only=True, 
                            monitor='val_loss', mode='min')
```

```python
result = model.fit(x_train, 
          y_train, 
          epochs=20, 
          batch_size=16, 
          validation_split=0.3,
          verbose=1,
          callbacks=[save_best])
```

```python
model.load_weights(filepath='yelp_comment.hdf')
eval_ = model.evaluate(x_test, y_test)
print(eval_[0], eval_[1]) # loss / accuracy
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def plot_model(result):
    acc = result.history['acc']
    val_acc = result.history['val_acc']
    loss = result.history['loss']
    val_loss = result.history['val_loss']
    x = range(1, len(acc)+1)
    
    plt.figure(figsize=(12, 5))
    plt.subplot(1,2,1)
    plt.plot(x, acc, 'b', label='Training acc')
    plt.plot(x, val_acc, 'r', label= 'Validation acc')
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='validation loss')
    plt.legend()
    
plot_model(result)
```
